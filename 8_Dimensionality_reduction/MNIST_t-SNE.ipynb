```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
plt.rcParams['figure.figsize'] = (16.0, 9.0)
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, cache=True)
```

```python
random_state=132
```

```python
X = mnist["data"]
Y = mnist["target"]
np.random.seed(random_state)
ids = np.random.permutation(60000)[:10000]
X = X[ids]
Y = Y[ids]
Y = Y.astype(np.int32)
```

```python
X_transformed = TSNE(n_components=2, random_state=random_state, n_jobs=-1).fit_transform(X)
X_transformed.shape
```

```python
def plot_plot(X,Y):
    plt.scatter(X[:,0],X[:,1], c=Y, cmap="jet", alpha=0.8)
    plt.colorbar()
    plt.show()
plot_plot(X_transformed,Y)
```

```python
from sklearn.svm import SVC
svc_clf = SVC()
```

```python
from sklearn.decomposition import PCA
```

```python
X_PCA = PCA(n_components=2).fit_transform(X)
```

```python
plot_plot(X_PCA,Y)
```

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
X_LLE = LocallyLinearEmbedding(n_components=2, n_jobs=-1).fit_transform(X)
plot_plot(X_LLE,Y)
```

```python
X_TNSE_3D = TSNE(n_components=3, random_state=random_state, n_jobs=-1).fit_transform(X)
X_TNSE_3D.shape
```

```python
from mpl_toolkits import mplot3d
%matplotlib notebook
```

```python
def plot_plot_3D(X,Y):
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    p = ax.scatter3D(X[:,0],X[:,1],X[:,2], c=Y, cmap="jet")
    fig.colorbar(p)
```

```python
plot_plot_3D(X_TNSE_3D,Y)
```
